- 3) Has GDP changed over time?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/all_data.csv')
df.head()
```

```python
df.info()
```

What Countries are represented in the dataset?

```python
df['Country'].unique()
```

Which Year are represented in the dataset?

```python
df['Year'].unique()
```

# Comparison of Life Expectancy and GDP Per Country

```python
df.groupby('Country')['GDP'].mean()
```

```python
ax = sns.barplot(x="Country", y="GDP", data=df)
plt.xticks(rotation=75)
plt.title('Comparison of the GDP On Various Countries');
```

The USA has the highest GDP while zimbabwe has the least GDP

```python
df.groupby('Country')['Life expectancy at birth (years)'].mean()
```

```python
ax = sns.barplot(x="Country", y="Life expectancy at birth (years)", data=df)
plt.xticks(rotation=45)
plt.title('Comparison of the Life Expectancy On Various Country');
```

- The life expectancy of most countries is around 70, with the least being Zimbabwe around 50

# Has GDP changed over time?

```python
sns.lineplot(x="Year", y="GDP", data=df)
plt.xticks(rotation=75)
plt.title('Comparison of the GDP Over the Years');
```

Based on the results: The GDP has increased over the years

# Comparison of GDP Over Years for the Different Countries

```python
ax = sns.pointplot(x="Year", y="GDP", hue = "Country", data=df)
plt.xticks(rotation=45)
plt.title('Comparison of The Various Countries GDP Over the Years');
```

Based on the results the Life expectancy of United States of America and China have an increased exponential rate over the years, While Chile, Germany, Zimbabwe, Mexico have a lower GDP overall that does not significantly increase over the years

# Has Life Expectancy changed over time?

```python
sns.lineplot(x="Year", y="Life expectancy at birth (years)", data=df)
plt.ylim(70,80)
plt.xticks(rotation=75)
plt.title('Comparison of the Life Expectancy Over the Years');
```

Based on the results: The life expectancy has increased over the years

```python
ax = sns.pointplot(x="Year", y="Life expectancy at birth (years)", hue = "Country", data=df)
plt.xticks(rotation=45)
plt.title('Comparison of The Various Countries Life Expectancy Over the Years');
```

Based on the results the Life expectancy of Zimbabwe have an increased exponential rate over the years, While the rest have a higher life expectancy but do not have a significant increase over the years
